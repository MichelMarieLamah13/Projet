# Variable tedCanId

Cet attribut répresente l'identifiant de l'avis d'attribution du lot

```python
tedCanId = lots['tedCanId']
```

## Nature de la variable

Cet attribut est un **attribut numérique** encodé sur un **entier** 

## Distribution

```python
tedCanId_vc = tedCanId.value_counts()
```

```python
tedCanId_vc
```

Pour étudier la distribution, nous allons nous baser sur le nombre de lots par avis d'attribution

Comme nous pouvons le voir, nous avons **410 283** avis d'attribution unique, du coup en répresentant directement ces valeurs, nous aurons un graphe illisible, pour éviter celà nous allons d'abord effectuer une discrétisation de ses valeurs pour les transformer en intervalles puis nous les représenterons sur un histogramme. 

```python
bins = [tedCanId_vc.min()-1, 1, 10, tedCanId_vc.max()]
tedCanId_bins = pd.cut(tedCanId_vc.values, bins=bins, precision=0)
```

```python
tedCanId_bins_vc = tedCanId_bins.value_counts()
```

```python
tedCanId_cut = pd.DataFrame({'Interval': tedCanId_bins_vc.index, 'CAN': tedCanId_bins_vc.values})
```

```python
tedCanId_cut['Proportion'] =  (tedCanId_cut['CAN']/ tedCanId_cut['CAN'].sum())*100
```

```python
tedCanId_cut
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_tedIdCAN_1(data, logy=True):
    fig = plt.figure(figsize=(8, 6))
    plt.xticks(fontsize=6)
    
    X = data['Interval'].values.astype(str)
    y = data['Proportion'].values
    
    # Utiliser une colormap pour les couleurs des barres
    colors = plt.cm.viridis(data["Proportion"]/100) 

    # Ajouter une barre de couleur (légende) pour indiquer les valeurs
    cbar = plt.colorbar(plt.cm.ScalarMappable(cmap='viridis'))
    cbar.set_label('Couleurs')
    
    plt.bar(X, y, color=colors)
    plt.xticks(rotation=0)
    if logy:
        plt.yscale('log')
    plt.title("Proportion d'avis d'attribution par nombre de lots")
    plt.xlabel('Nombre de Lots')
    plt.ylabel("Proportion d'avis d'attribution")
```

```python
plot_tedIdCAN_1(tedCanId_cut)
```

Nous pouvons essayer de representer les proportions d'avis d'attributions ayant exactement un lot et ceux ayant plus d'un lot

```python
tedCanId_vc_1 = tedCanId_vc[tedCanId_vc == 1]
```

```python
tedCanId_vc_other = tedCanId_vc[tedCanId_vc > 1]
```

```python
data = pd.DataFrame({
    "Lots": ["1 Lot", "Plus d'un lot"],
    "Avis d'attribution": [len(tedCanId_vc_1), len(tedCanId_vc_other)]
})
data["Proportion"] = (data["Avis d'attribution"]/len(tedCanId_vc))*100
```

```python
data
```

Nous pouvons remarquer que **240 176** (59%) des avis d'attributions ont exactement **1 lot** et 170 107 (**41%**) plus d'un lot.

Graphiquement, on aura:
